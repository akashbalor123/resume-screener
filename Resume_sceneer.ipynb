Resume Loader

```python
def load_resume(resume_input_path):
    try:
        file_data = []
        for file in os.listdir(resume_input_path):
            if file.endswith('.txt'):
                file_path = os.path.join(resume_input_path, file)
                res_id, cad_id = file.split('_', 1) 
                res_id = int(res_id)
                cad_id = cad_id.replace('.txt', '')
                with open(file_path, 'r') as f:
                    file_text = f.read()
                file_data.append({'resume_id': res_id, 'candidate_id': cad_id, 'resume_text': file_text})
        df= pd.DataFrame(file_data)
        df = df.sort_values(by='resume_id').reset_index(drop=True)
        return df
    except Exception as e:
        print(f"Error loading resumes: {e}")
        return pd.DataFrame()
```

```python
load_resume(resume_input_path)
```
